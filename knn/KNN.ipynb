# KNN(K-近邻算法) 
分类是投票 回归是均值
如果某个样本在特征空间中的K个最相似（样本距离越近越相似）的样本中的大多数（分类）属于一个类别，则该样本也属于这个类别
K值过小：容易受到异常点的影响，模型变得复杂容易过拟合
K值过大：收到样本均衡的影响，整体模型变得简单，欠拟合
欧式距离：对应维度差值平方和，开平方根

分类是投票 回归是均值
相同点：有监督学习（有特征，有标签）
不同点：分类（标签不连续）、回归（标签连续）

分类：
1.计算未知样本到每一个训练样本的距离
2.将训练样本按距离大小升序排列    升序排列
3.取出距离最近的K个训练样本
4.进行多数表决（投票），统计K个样本中哪个类别的样本个数最多
5.将未知样本归属到出现次数最多的类别


回归：
1.计算未知样本到每一个训练样本的距离
2.将训练样本按距离大小升序排列
3.取出距离最近的K个训练样本
4.把这K个样本的目标值计算平均值（均值）
5.作为未知样本预测的值



## 常用的距离度量方式

### 1.欧氏距离
对应维度差值平方和，开平方根

### 2.曼哈顿距离
也称"城市街区距离"，对应维度差值的绝对值，求和

### 3.切比雪夫距离 (可以斜着走)
对应维度差值的绝对值，求最大值

## 特征值预处理
特征的单位或大小相差较大，或是某特征的方差相比其他特征要大出几个数量级，容易影响目标结果

### 1.归一化
通过对原始数据进行变换把数据映射到[min,max]之间（默认[0,1]）,容易受到最大值和最小值的影响（鲁棒性较差），所以一般用于处理精确小数据场景
$$
X' = \frac{x-min}{max - min}
$$
$$
X'' = X' * (max - min) + min
$$

### 2.标准化
通过对原始数据进行标准化，转换为均值为0，标准差为1的标准正态分布的数据，mean为特征的均值，σ是特征的标准差
少量异常点对于具有一定数据量的数据集的平均值影响不大，适合现代嘈杂大数据场景
$$
X' = \frac{x-mean}{σ}
$$


## 交叉验证
是一种划分数据集的方法，目的是为了得到更加准确可信的模型评分

### 交叉验证原理
n折交叉验证：将数据集分为n份，，分别对不同参数k进行模型训练
1.第一次：把第一份数据作为验证集，其他数据作为训练集
2.第二次：把第二份数据作为验证集，其他数据作为训练集
3.以此类推，一共训练n次，评估n次
4.取平均值做交叉验证，选出一个最优的模型超参数k
5.若k = i时模型得分最好，则再使用全部的训练集（训练集+验证集）对k=i模型进行训练，
再用测试集对k=i模型进行评估


## 网格搜索
将若干参数传递给网络搜索对象，对每组超参数都采用交叉验证评估，得到一组最优的超参数
